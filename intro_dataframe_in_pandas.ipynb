#### type casting

```python
df['Number']=df['Number'].astype(int).astype(str)
```

```python
df.describe()
```

```python
df['Position']=df['Position'].astype('category')
```

```python
df.info()
```

```python
def turn_height(height):
    f,i=height.split('-')
    return int(f)*30.48+int(i)*2.54
```

```python
df['Height'].apply(lambda x:int(x.split('-')[0])*30.48+int(x.split('-')[1])*2.54)
```

```python
df['Height']=df['Height'].apply(turn_height)
```
