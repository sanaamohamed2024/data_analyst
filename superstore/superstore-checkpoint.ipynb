#### 1. Parsing and Merging:

```python
def con_yearly_data(file_paths, output_file):
   #file_paths the list that contain all the file paths
   #output_file the file where we store the data 
    all_data = []  # List to store data from each file
    
    for file_path in file_paths:
        # Load and parse each file
        df = pd.read_csv(file_path)
        all_data.append(df)
    
    # Concatenate all data into a single DataFrame
    con_data = pd.concat(all_data, ignore_index=True)
    
    # Save the consolidated DataFrame to a CSV file
    con_data.to_csv(output_file, index=False)
    print(f"Data successfully concatenate into {output_file}")

```

#### 2. Data Cleansing:

```python
#Removes duplicate rows that match the column headers in the DataFrame.

# Get column headers as a list
headers = df.columns.tolist()

# Check each row: keep only rows that are NOT identical to the headers
df = df[~df.apply(lambda row: list(row) == headers, axis=1)]
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Cast each column to the appropriate data type. All columns are initially loaded as object types in pandas
df['Order ID']=df['Order ID'].astype(str)
df['Quantity Ordered']=df['Quantity Ordered'].astype(int)
df['Product']=df['Product'].astype(str)
df['Purchase Address']=df['Purchase Address'].astype(str)
df['Price Each']=df['Price Each'].astype(float)
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df.info()
```

```python
df.describe()
```

```python
# Filter the data to include only records from 2019
df = df[df['Order Date'].dt.year == 2019]
```

#### 3. Data Transformation:

```python
# Extract City and State information from the "Purchase Address" column and create two new columns for them

# Split the "Purchase Address" into components
df[['City', 'State']] = df['Purchase Address'].str.extract(r',\s*([^,]+),\s*([A-Z]{2})\s*\d{5}')

```

```python
df.head()
```

```python
# Identify the granularity of the data. Note that an order may span multiple rows if it contains
#different products. Consider each unique "Order ID" as a single order.
# Step 1: Calculate total price for each product in each row (Price Each * Quantity)
df['Total Price'] = df['Price Each'] * df['Quantity Ordered']

# Step 2: Group by 'Order ID' and aggregate data
# We'll sum the 'Total Price', count the number of products, and list the product names
order_summary = df.groupby('Order ID').agg(
    Total_Sales=('Total Price', 'sum'),  # Sum of all product prices for each order
    Product_Count=('Product', 'count'),  # Count the number of products in each order
    Products=('Product', lambda x: ', '.join(x))  # List all product names in each order
).reset_index()
```

#### 2. Trend Analysis:

```python
# Extract the day of the week from the Order Date
df['Day of Week'] = df['Order Date'].dt.day_name()

# Count the number of unique orders for each day of the week
orders_per_day = df.groupby('Day of Week')['Order ID'].nunique()

# Reorder the days of the week for logical ordering
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
orders_per_day = orders_per_day.reindex(days_order)

# Plot the results
plt.figure(figsize=(10, 6))
bars = plt.bar(orders_per_day.index, orders_per_day.values, color='skyblue', edgecolor='black')

# Add text annotations above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', ha='center', va='bottom', fontsize=10)

# Add titles and labels
plt.title('Number of Orders by Day of the Week', fontsize=16)
plt.xlabel('Day of the Week', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Extract month from Order Date
df['Month'] = df['Order Date'].dt.month_name()

# Count the number of unique orders for each month
orders_per_month = df.groupby('Month')['Order ID'].nunique()

# Reorder months for chronological order
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
orders_per_month = orders_per_month.reindex(month_order)

# Plot the results
plt.figure(figsize=(12, 6))
orders_per_month.plot(kind='line', marker='o', color='blue', linewidth=2)

# Annotate the data points
for i, value in enumerate(orders_per_month.values):
    plt.text(i, value + 10, f'{value}', ha='center', fontsize=10)

# Add titles and labels
plt.title('Preferred Months for Customer Orders', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Orders', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

#### 4. Category Analysis:

```python
# Extract year and month for time-based analysis
df['Year-Month'] = df['Order Date'].dt.to_period('M')

# Group by category and year-month, then aggregate the total sales and revenue
category_sales = df.groupby(['Category', 'Year-Month']).agg(
    total_sales=('Quantity Ordered', 'sum'),
    total_revenue=('Total Price', 'sum')
).reset_index()

# Create a line plot to visualize trends in sales and revenue by category
plt.figure(figsize=(12, 8))

# Plot total sales for each category over time
plt.subplot(2, 1, 1)
for category in category_sales['Category'].unique():
    category_data = category_sales[category_sales['Category'] == category]
    plt.plot(category_data['Year-Month'].astype(str), category_data['total_sales'], label=category)

plt.title('Product Category Sales Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.legend()

# Plot total revenue for each category over time
plt.subplot(2, 1, 2)
for category in category_sales['Category'].unique():
    category_data = category_sales[category_sales['Category'] == category]
    plt.plot(category_data['Year-Month'].astype(str), category_data['total_revenue'], label=category)

plt.title('Product Category Revenue Over Time')
plt.xlabel('Year-Month')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
