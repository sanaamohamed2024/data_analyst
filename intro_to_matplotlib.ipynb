```python
import matplotlib.pyplot as plt
import numpy as np
```

### supported visuals

#### line plot it uses to plot functions

```python
plt.figure(figsize=(15,6))
plt.suptitle('X vs Y')
plt.subplot(1,2,1)
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.plot(x,y,marker='o',color='green',linestyle='dashed',linewidth='3')
plt.xlim([0,10])
plt.ylim([0,10])
plt.legend(['T shirt sales'])
plt.title("x vs y")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.subplot(1,2,2)
x2=np.array([2,4,6,8])
y2=np.array([1,2,3,4])
plt.plot(x2,y2)
plt.grid()
plt.title("x2 vs y2")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.axvline(5, linestyle='dashed')  # to plot vertical line 
plt.axhline(4, linestyle='dotted')  # to plot herozontal line 
plt.legend(['short sales','sales change','target'],loc='lower right') # we use loc to choose the position to write legend
plt.xticks(list(range(11)),rotation=90) # to change in the steps and the shape of numbers in x axis
plt.yticks(np.arange(0,12,1))
plt.savefig('x vs y') # if you want to save fig it must be before plt.show
plt.tight_layout() # if we have the problem of interfaces between the showing of fig it must be in the end and before plt.show()
plt.show()
```

```python
import pandas as pd
df=pd.read_csv('Downloads/Superstore.csv')
df.columns

```

```python
df.groupby('Order Date')['Order ID'].nunique()
```

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
```

```python
df.groupby([df['Order Date'].dt.year,df['Order Date'].dt.month])['Order ID'].nunique()
```

```python
to_draw_1=df.resample('M',on='Order Date')['Order ID'].nunique()
```

```python
to_draw_2=df.resample('Y',on='Order Date')['Order ID'].nunique()
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
x=to_draw_1.index
y=to_draw_1
plt.plot(x,y,marker='o',color='green',linestyle='dashed',linewidth='3')
plt.title("unique order for month")
plt.xlabel("Date")
plt.ylabel("number of order")
plt.grid()
plt.subplot(1,2,2)
x2=to_draw_2.index
y2=to_draw_2
plt.plot(x2,y2)
plt.grid()
plt.title("unique order for year")
plt.xlabel("Date")
plt.ylabel("number of order")
plt.tight_layout() # if we have the problem of interfaces between the showing of fig it must be in the end and before plt.show()
plt.show()
```

#### bar plotn

```python
x=['cairo','alex','aswan']
y=[100,250,140]
colors=['red','blue','green']
bars=plt.bar(x,y,color=colors)
for i,v in enumerate(bars):
    plt.text(v.get_x()+v.get_width()/2, v.get_height()+5,int(v.get_height()),va='center',ha='center',fontsize=20)
```

```python
to_draw_1=df['State'].value_counts().drop_duplicates()[:10]
```

```python
to_draw_2=df['State'].value_counts().drop_duplicates()[-10:]
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1,2,1)
x=to_draw_1.index
y=to_draw_1
bars=plt.bar(x,y,color='red')
for i,v in enumerate(bars):
    plt.text(v.get_x()+v.get_width()/2, v.get_height()+100,int(v.get_height()),va='center',ha='center',fontsize=20)
plt.xticks(rotation=90)    
plt.subplot(1,2,2)
x2=to_draw_2.index
y2=to_draw_2
bars=plt.bar(x2,y2,color='blue') # barh to plot the line herizontal 
for i,v in enumerate(bars):
    plt.text(v.get_x()+v.get_width()/2, v.get_height()+1,int(v.get_height()),va='center',ha='center',fontsize=20)
plt.xticks(rotation=90)    
plt.tight_layout() # if we have the problem of interfaces between the showing of fig it must be in the end and before plt.show()
plt.show()
```

```python
plt.figure(figsize=(15,6))
for mode in df['Ship Mode'].unique():
    to_draw=df[df['Ship Mode']==mode].drop_duplicates(['Order ID']).resample('M',on='Order Date')['Order ID'].nunique()
    plt.plot(to_draw.index,to_draw)
    plt.title("Order count per ship mode")
    plt.xlabel("Date")
    plt.ylabel("number of order")
    plt.legend(df['Ship Mode'].unique())
plt.show()    
   
```

#### pie chart

```python
to_draw=df['Ship Mode'].value_counts()
plt.pie(to_draw,labels=to_draw.index,autopct='%1.2f%%') # 1.2f the 2 refer to the first two number after decimal mark
plt.show()
```
